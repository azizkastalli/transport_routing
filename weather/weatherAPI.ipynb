```python
import csv
import codecs
import urllib.request
import pandas as pd
import datetime
```

```python
def date_format(year,month,day):
    date = year+'-'+month+'-'+day+'T00%3A00%3A00'
    return date
```

```python
#aziz.kastalli02@gmail.com 9RWG9ACQYHPT5S6L5M99XSCM8
#aziz.kastalli@esprit.tn 8UXGT5G59YEGQ6KKPHQQ99SRE

def build_request(start_DateTime, end_DateTime, location='Tunis', quertype = 'history',
                  key = '8UXGT5G59YEGQ6KKPHQQ99SRE', content_type = 'csv'):

    URL_Request = 'https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/'+\
                  'weatherdata/'+quertype+\
                  '?aggregateHours=1'+\
                  '&combinationMethod=aggregate'+\
                  '&startDateTime='+start_DateTime+\
                  '&endDateTime='+end_DateTime+\
                  '&maxStations=-1'+\
                  '&maxDistance=-1'+\
                  '&contentType='+content_type+\
                  '&unitGroup=metric'+\
                  '&locationMode=single'+\
                  '&key='+key+\
                  '&dataElements=default'+\
                  '&locations='+location
    return URL_Request
```

```python
def transform_request(CSVText,weather_data):
    col_names = True
    rows = 0
    for Row in CSVText:
        rows+=1
        if col_names :
            #skip names row
            col_names = False
        else:
            weather_data.append(Row)
    return weather_data,rows
```

```python
columns  = ['Name','Date time','Maximum Temperature','Minimum Temperature','Temperature',
            'Wind Chill','Heat Index','Precipitation','Snow Depth','Wind Speed',
            'Wind Gust','Visibility','Cloud Cover','Relative Humidity','Weather Type',
            'Conditions']

weather = pd.DataFrame(columns=columns)
```

```python
date_start = datetime.datetime(2021,1,14,0,0,0)
date_end = datetime.datetime(2021,1,18,0,0,0)
stop = datetime.datetime(2021,3,30,0,0,0)
weather_data = []
total_rows = 0

while date_end<stop:
    start_DateTime = date_start.isoformat().replace(':','%3A')
    end_DateTime   = date_end.isoformat().replace(':','%3A')
    print(start_DateTime, '  ', end_DateTime)
    #--------------request-----------------
    URL = build_request(start_DateTime,end_DateTime)
    CSVBytes = urllib.request.urlopen(URL)
    CSVText = csv.reader(codecs.iterdecode(CSVBytes, 'utf-8'))
    weather_data,rows = transform_request(CSVText,weather_data)
    total_rows += rows
    if total_rows-rows == 1:
        print('لقد نفذ رصيدكم')
        break
    print('total requests : ',total_rows)
    #--------------request-----------------    
    date_end   += datetime.timedelta(days=4)    
    date_start += datetime.timedelta(days=4)
```

```python
weather = pd.DataFrame(weather_data , columns=columns)
weather.tail()
```
